```python
import mne
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use('Qt5Agg')
# Load a single UCSD EEF file
raw = mne.io.read_raw_bdf('UCSanDiego\sub-hc1\ses-hc\eeg\sub-hc1_ses-hc_task-rest_eeg.bdf', preload=True)
print(raw.info)
# preload=True loads data into memory
```

```python
raw.plot(scalings='auto', title='Raw EEG', show=True, block=True)

# Keep only EEG channels
raw.pick(picks='eeg')
```

```python
psd = raw.compute_psd(fmin=1, fmax=40)  # returns a PSD object

# Plot PSD
psd.plot()
```

```python
import numpy as np
data = raw.get_data()
print(data)
```
