```python
import mne
import matplotlib 
import matplotlib.pyplot as plt
matplotlib.use('Qt5Agg')
from mne.preprocessing import ICA
from mne.preprocessing import read_ica
raw = mne.io.read_raw_fif('rEEG\sub-001\preICA_raw.fif', preload=True)
ica = read_ica(f'rEEG\sub-001\ica-001-variance_compica.fif')
```

```python
bad_ics=[0,6,2,20,32,33,30]
ica.exclude=bad_ics
clean_raw = ica.apply(raw.copy())
```

```python
# # --- Plot before ICA cleaning ---
# raw.plot( title='Original EEG (Before ICA)')

# # --- Plot after ICA cleaning ---
# clean_raw.plot( title='Cleaned EEG (After ICA)')


# Pick a representative channel, e.g., Cz
picks = ['F3', 'T8']

# Compute PSDs
psd_orig = raw.compute_psd(picks=picks, fmin=1, fmax=50, n_fft=2048)
psd_clean = clean_raw.compute_psd(picks=picks, fmin=1, fmax=50, n_fft=2048)
freqs = psd_orig.freqs  # same for both

plt.figure(figsize=(10,6))
for i, ch in enumerate(picks):
    plt.semilogy(freqs, psd_orig.get_data()[i, :], label=f'{ch} Original')
    plt.semilogy(freqs, psd_clean.get_data()[i, :], label=f'{ch} Cleaned', linestyle='--')

plt.xlabel('Frequency (Hz)')
plt.ylabel('PSD (uVÂ²/Hz)')
plt.title('EEG PSD Before and After ICA Cleaning')
plt.legend()
plt.show()
```
