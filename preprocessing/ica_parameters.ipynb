```python
import mne
import matplotlib 
import matplotlib.pyplot as plt
matplotlib.use('Qt5Agg')

raw = mne.io.read_raw_fif('rEEG\sub-128\preICA128_raw.fif', preload=True)
raw.plot()
```

```python
import numpy as np

# basic info
print(raw.info['nchan'], len(raw.info['bads']))

# channel variances (zero/near-zero channels reduce rank)
data = raw.get_data(picks='eeg')
vars = np.var(data, axis=1)
print("min/median/max var:", vars.min(), np.median(vars), vars.max())
print("channels with tiny var:", np.where(vars < 1e-12)[0])

# SVD-based rank estimate (robust)
u, s, vh = np.linalg.svd(data - data.mean(axis=1, keepdims=True), full_matrices=False)
rank_est = np.sum(s > (s.max() * 1e-6))   # threshold can be adjusted
print("estimated rank (SVD):", rank_est)

# covariance rank
cov_rank = np.linalg.matrix_rank(np.cov(data))
print("covariance rank:", cov_rank)
```

```python
from mne.preprocessing import ICA
ica = ICA(
    n_components=.95,
    method='infomax',
    fit_params=dict(extended=True),
    random_state=42
)

ica.fit(raw)
```

```python
ica.plot_components()
for comp in [0,3,6]:
    ica.plot_properties(raw, picks=comp)
```
