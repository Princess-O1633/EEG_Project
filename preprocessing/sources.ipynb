```python
import mne
import os
import matplotlib
matplotlib.use('Qt5Agg')
import matplotlib.pyplot as plt
from mne.preprocessing import read_ica

# -------------------------
# CONFIG
# -------------------------
BASE_DIR = r"rEEG"
SUBJECTS = [f"sub-{i:03d}" for i in range(45, 48)]

RAW_NAME = "preICA{sid}_raw.fif"
ICA_NAME = "ica-{sid}-variance_compica.fif"
OUT_NAME = "postICA{sid}_raw.fif"

# -------------------------
# LOOP
# -------------------------
for subj in SUBJECTS:
    sid = subj.split("-")[1]
    print(f"\n=== Processing {subj} ===")

    raw_path = os.path.join(BASE_DIR, subj, RAW_NAME.format(sid=sid))
    ica_path = os.path.join(BASE_DIR, subj, ICA_NAME.format(sid=sid))
    out_path = os.path.join(BASE_DIR, subj, OUT_NAME.format(sid=sid))

    if not os.path.exists(raw_path) or not os.path.exists(ica_path):
        print("Missing file — skipping")
        continue

    # Load raw and ICA
    raw = mne.io.read_raw_fif(raw_path, preload=True)
    ica = read_ica(ica_path)

    # Optional visualization (can comment out for speed)
    # ica.plot_sources(raw, show_scrollbars=True)
    plt.show()

    # Manual IC entry (from Notion)
    reject_str = input(
        f"Enter ICs to reject for {subj} (comma-separated, e.g. 0,1,4) or press Enter for none: "
    )

    if reject_str.strip():
        # Remove duplicates and sort
        ica.exclude = sorted(set(int(x.strip()) for x in reject_str.split(",")))
    else:
        ica.exclude = []

    # Apply ICA
    raw_clean = raw.copy()
    ica.apply(raw_clean)

    # Save cleaned raw
    raw_clean.save(out_path, overwrite=True)
    print(f"Saved cleaned raw → {out_path}")

print("\n=== ALL SUBJECTS PROCESSED ===")
```
