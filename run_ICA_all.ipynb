```python
import mne
import matplotlib 
import matplotlib.pyplot as plt
matplotlib.use('Qt5Agg')
from mne.preprocessing import ICA

for i in range(9, 10):
    raw = mne.io.read_raw_fif(f'rEEG\\sub-{i:03d}\\preICA_raw.fif', preload=True)

    # Fit ICA with fixed number of components
    ica = ICA(n_components=42, method='infomax', random_state=42)
    ica.fit(raw)

    # Save ICA object
    ica.save(f'rEEG/sub-{i:03d}/ica-{i:03d}-variance_compica.fif', overwrite = True)

```
